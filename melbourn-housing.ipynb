```python
num_cols = df.select_dtypes(include=np.number).columns.drop("Price").to_list()
cat_cols = [col for col in df.columns.to_list() if col not in num_cols]
```

```python
# Plot numeric columns distributions

plt.figure(figsize=(18,16))
for i,col in enumerate(num_cols):
    plt.subplot(4, 3, i+1)
    sns.histplot(x=col, data=df, bins=40)
    plt.title(col)
    plt.ylabel(" ")
    plt.xlabel(" ")
```

```python
cat_cols = [col for col in cat_cols if col not in ('Address', 'Suburb', 'Date', 'SellerG', 'Price')]
```

```python
# Plot categoric columns distributions

plt.figure(figsize=(18,16))
for i,col in enumerate(cat_cols):
    plt.subplot(2, 2, i+1)
    sns.countplot(y=col, data=df)
    plt.title(col)
    plt.ylabel(" ")
    plt.xlabel(" ")
```

```python
corr = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, vmin=0, vmax=1, linewidths=2, square=True, cmap="Blues")
```
